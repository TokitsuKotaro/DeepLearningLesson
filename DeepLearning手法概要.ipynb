# 強化学習

## 全体像

image.png

## 全体観

強化学習のモデルには、価値ベースと方策ベースがある。

価値ベースは、その状況における行動の価値を学習するモデル

- 収束速度が早い、内容を理解しやすい。
- 行動空間が広いと学習が難しい。メモリの消費が激しい。

方策ベースは、その状況における最適な行動を直接学習するモデル

- 行動空間が広くても学習できる。
- 収束速度が遅い。局所解になる可能性がある。

という特徴がある。

本資料で扱うのは価値ベースのモデルのみ。

## Q学習

Q学習の図形的説明

image.png

Q値を用いた強化学習

Q値とは「その状態での即時的に得られる報酬の価値」と

「見積もられた次の状態の価値」の和

- Q値が無いとき

 その時の最善手を取るので、テトリスを細かく消してしまう

- Q値が有るとき

 その後の状態を予測して動くので、テトリスのブロックを消さずにためて、まとめて消す

## DQN

DQNのネットワーク構造

image.png

Q値の学習をDeepLearningによって行うモデル

- モデルサイズが大きく、メモリの消費が激しい。
- 正確にQ値を推定することが出来るため、学習精度が高い

## 経験再生

行動と結果を経験として記録しておき、後にその過去の経験をサンプリングして学習データにする方法。

逐次的に学習を進めると、時系列的な相関が発生してしまう。

経験再生は、定期的に時間的に離れた過去のデータを使用することで、時系列的な相関が生まれるのを防ぐ

**優先順位付き経験再生**

サンプルに優先順位を付けて使用する方法を取ることが多い。

- TD誤差が大きくなるサンプルからサンプリングし、Q学習ではTD誤差が小さくなるようにする。

 →DQNでも用いられる

## 分散学習

分散学習のネットワーク構造

image.png

複数のプレイヤー(Actor)を用意し、それぞれゲームをプレイさせる方法

プレイ結果は記憶として親(Learner)に引き継がれ

親はそれを使って良い行動を学習する。

学習結果は、一定期間ごとにプレイヤー(Actor)にも教えられ、学習が進む。

並列してゲームをプレイ出来るので、学習時間を大幅に短縮することが出来る。

## Dueling Network

Q学習とDWの挙動比較図

image.png

有名なCartPole(倒立振子)問題を元に説明する。

振子が倒れかけのときに、カートを右に押そうが
左に押そうが、結果は変わらない。

これはカートの状態に依存した部分と言える。

一方で、カートに何らかの行動をして、結果が大きく変わることもある。

これは行動に依存した部分と言える

通常のQ学習ではこれらはまとめて計算されていたが、DuelingNetworkはこれらを分離して考える。

これにより、学習時間の短縮と精度向上を実現
